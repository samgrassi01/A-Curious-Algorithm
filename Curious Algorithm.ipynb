# A Curious Algorithm

Consider the following simple algorithm:  

1. Given any triangle, $P$, pick a point anywhere along one of $P$'s edges and call it `current_point`.  
2. From `current_point`, draw to the *farthest* edge of $P$, such that the line drawn is perpindicular to the chosen edge.
3. Set `current_point` equal to the end of this drawn perpindicular line, and repeat the process for some finite number of steps.

Following this "right angle algorithm" on paper, I noticed a couple curious properties, the most prominent of which was that given any triangle and any starting point, the algorithm would quickly converge to some repeating shape, which I will call the "ultimate shape".  

Interested in finding exact properties of this ultimate shape, I wrote a program to generate a triangle and implement this right angle algorithm, graphing the result.  I imported the algorithm into this notebook for use, hiding the code used to build it.  If you are interested in seeing the code, refer to the `TriangleFunctions.py` file in the GitHub repo.

An implementation of this algorithm and subsequent analysis of its more interesting results follows below.

## Putting the Algorithm to the Test

Lets start by showing what the algorithm comes up with when passed a simple equilateral triangle.

**NOTE:** The function `return_points` takes two angles and the right point of the triangle to build, returning the necessary points to make the triangle with the passed exact angles and right end point.  The function `drawTriangle` takes `(number_of_iterations, starting_point, outer_triangle_points, steps_to_show)` where `number_of_iterations` is the number of steps the algorithm will take, `starting_point` is where the algorithm will start drawing from, `outer_triangle_points` is the set of three points generated by `return_points`, and finally `steps_to_show` is the index-range of steps to graph.

```python
eq_triangle = return_points(60, 60, [4, 0])
visited_points = drawTriangle(100, [3.4, 0], eq_triangle, [0, 100])
```

As you can see, in the case of this equilateral triangle with the algorithm starting at point $(3.4, 0)$, after just a few steps the algorithm converged into a different triangle within the outer one.  This inner ultimate triangle has each edge perpindicular to one of the outer triangles edges, with all three of its vertices touching the outer triangles edges.

Lets use another function, `give_info`, to eak some more information out of this.  

```python
give_info(visited_points, eq_triangle)
```

$\space$

As you can see, the ultimate inner triangle is *similar* to the outer one, as all the inner angles are the same!  A rather interesting result.  

With the first steps of the process not graphed, you can clearly see the ultimate triangle within the outer one in the graph below, and see that it is indeed an equilateral triangle!

```python
visited_points = drawTriangle(100, [3.4, 0], eq_triangle, [90, 100])
```

Now, lets take a look at what happens when the algorithm is applied to an isosceles triangle.  The ultimate inner triangle angles and outer triangle angles are printed out along with the resultant graph.

```python
iso1 = return_points(50, 65, [4, 0])
visited_points = drawTriangle(100, [3.6, 0], iso1, [0, 100])
give_info(visited_points, iso1)
```

As you can see, the same thing happened for this particular isosceles triangle!  The ultimate triangle that the algorithm converged to is similar to the outer triangle!

At this point, you are likely wondering what would happen as we fiddle with the top angle and starting point of the algorithm, as the ultimate shape may not always be a triangle!  I will get to that, but first I am going to show that this property of convergence to similar triangles also holds for scalene triangles.

```python
scalene1 = return_points(60, 70, [4, 0])
visited_points = drawTriangle(100, [3.6, 0], scalene1, [0, 100])
give_info(visited_points, scalene1)
```

Once again, we can see the inner ultimate triangle is similar to the outer triangle.  

One following natural question may be, "If the computational resources are not available to find this ultimate triangle, how else might it be found exactly"?  

The answer would be that this exact inner triangle could be found with relative ease using geometry.  Take a look at the general diagram below for one of these cases.

<img src="General_triangle.png" width=700px>

We know $a, b, c, A, B, C$, we also know that each of the edges of the inner triangle will be perpindicular with its corresponding edge on the outer triangle, and that the inner ultimate triangle will be similar to the outer one (same angles).  With this information, all the inner angles can be found, giving enough information to solve for $x, y$, and  $z$ shown in the diagram above.  After finding these values, the points of the inner ultimate triangle can be found.  

One interesting formula that can be used to *approximate* the side lengths of the inner ultimate triangle is given by

$$
a' \approx \frac{sin(\gamma) \cdot a}{\varphi}
$$

where $\gamma$ is the top angle of the outer triangle, $a'$ is the inner side length to be approximated, $a$ is the corresponding side length, and $\varphi=1.61803 \cdots$, the Golden Ratio.  

The side length to be solved for, $a'$, can be found exactly with

$$
a' = \frac{sin(\gamma) \cdot a \cdot c}{\varphi}
$$

where $c=1\pm \eta$ where $\eta$ is some very small constant unique to the outer triangle.  I have a hunch $\eta$ is given by a function of the top angle, however I have yet to find this function exactly.

## Other Ultimate Shapes

As alluded to earlier, when the starting point of the algorithm or the angles of the outer triangles are adjusted, the algorithm generates other interesting repeating shapes.  One of the simpler repeating shapes is given below from an isosceles triangle.

**Note** for these new ultimate shapes below, I will only be graphing the first steps of the algorithmic process when I feel that showing those first steps is a value add.  In most of the below cases this is not the case, so I will do away with them.

```python
iso2 = return_points(30, 75, [4, 0])
visited_points = drawTriangle(100, [3.6, 0], iso2, [90, 100])
```

This is the first repeating ultimate shape we have seen that produces something other than a triangle.  The above shape is nothing to get to excited about however so lets start by slightly increasing and decreasing the top angle of the triangle and see what else the algorithm comes up with:

```python
iso3 = return_points(40, 70, [4, 0])
visited_points = drawTriangle(100, [2, 0], iso3, [90, 100])
```

```python
iso4 = return_points(20, 80, [4, 0])
visited_points = drawTriangle(100, [3, 0], iso4, [80, 100])
```

Those patterns are much more interesting! Recall that the algorithm will always draw the longest possible line it can while still obeying the right angle rule, as we can see the result of this taking form in the previous two examples.  For this specific isosceles triangle, as the top angle decreases, the number of "zig-zags" also increases.  

As shown below, scalene triangles succumb to a similar phenomena as well, producing different repeating shapes.

```python
scalene2 = return_points(40, 80, [4, 0])
visited_points = drawTriangle(110, [3, 0], scalene2, [90, 110])
```

```python
scalene3 = return_points(20, 86, [4, 0])
visited_points = drawTriangle(110, [2, 0], scalene3, [80, 110])
```

Lets take a look at what happens when we move the top angle in the other direction, starting with a $45-90-45$ triangle.  In this case, it is interesting seeing all the steps the algorithm takes, so I plot them all below.  The starting point is $(3.7, 0)$.

```python
iso6 = return_points(90, 45, [4, 0])
visited_points = drawTriangle(130, [3.7, 0], iso6, [0, 130])
```

The final ultimate shape is

```python
visited_points = drawTriangle(110, [3, 0], iso6, [100, 110])
give_info(visited_points, iso6)
```

When the algorithm was implemented on the above right triangle, it produced some interesting results.  The most immidiately noticable of which being that the hourglass like ultimate shape, is built out of two $45-90-45$ triangles!  In fact, we will soon see that this is the case for all obtuse triangles!

For context into the algorithms decisions above, in (almost) all cases, if it is presented with two equal distances it could follow (that when drawn will form a perpindicular line), it will randomely choose one of them to follow.  However, in the current implementation of the algorithm I have prevented it from drawing along edges or meeting the triangles vertices to force it into developing the hourglass ultimate shape seen above.  It is for this reason that on what looks to be the fourth step above, the algorithm chose to draw down instead of up to the top vertice of the triangle.  If the algorithm had been allowed to do so, the resulting inner shape would have been the two ultimate triangles flipped around the y-axis and scaled up, as pictured below by a previous implementation of the algorithm.

Below, a few more obtuse triangles will be implemented to show this property of an "hour-glass" ultimate shape composed to two similar triangles holds for all obtuse triangles (both scalene and isosceles).

```python
iso7 = return_points(100, 40, [4, 0])
visited_points = drawTriangle(150, [3.4, 0], iso7, [100, 150])
give_info(visited_points, iso7)
```

```python
scalene4 = return_points(110, 30, [4, 0])
visited_points = drawTriangle(150, [1, 0], scalene4, [100, 150])
```

```python
iso8 = return_points(140, 20, [4, 0])
visited_points = drawTriangle(150, [2.2, 0], iso8, [100, 150])
```

While this property is true for all obtuse triangles, note what happens as $\gamma$, the top angle, approaches 180 degrees.  As $\gamma \rightarrow 180$, the range of starting points at which the algorithm will converge to this ultimate hour-glass shape decreases, approaching the center of the triangle.  

Notice what happens if the algorithm starts outside of this "critical range".

```python
iso9 = return_points(140, 20, [4, 0])
visited_points = drawTriangle(100, [2.3, 0], iso9, [0, 100])
```

The algorithm gets "stuck" in a corner and approaches the left/right endpoints as the number of steps approaches infinity!

## Conjectures

First, we can make some conclusions about the "stability" of different classes of triangles as they relate to the simple right angle algorithm.  A class's "stability" can be treated as a function of the number of possible unique families of repeating shapes the algorithm will produce when applied to a particular class of triangle.  The more possible families of patterns produced, the lower the stability of the class of triangles, as it is harder to predict what particular pattern the algorithm will converge to when passed a triangle class of lower stability (scalene triangles for example).

1. The most stable triangle is the equilateral triangle.  Regardless of the equilateral triangle's size, the algorithm will always converge to drawing a single similar ultimate triangle.
2. The second most stable triangle type is the isosceles triangle.  The algorithm implemented on isosceles triangles may converge to one of just a few different families of repeating patterns, with the alternate option being the algorithm gets stuck in a corner.
3. The least stable triangle is the scalene triangle.  When the algorithm is used on a scalene triangle, there are many (I am not sure how many) different kinds of repeating patterns the algorithm will fall into.  The convergence of the algorithm on scalene triangles is also highely dependent on the starting point of the algorithm, and may go in many different directions based on the starting point alone.

**Conjecture:** for any given triangle $P$, there *may* exists *a single* similar triangle $P'$, such that when $P'$ is placed within $P$, $P'$ can touch all three of $P$'s edges, and will form a right angle with each of $P$'s edges.  Such a triangle $P'$ does not exist for triangles with an angle greater than or equal to $90$ degrees, and does not exist for specific accute triangles.  A $P'$ does exist however for every equilateral triangle.  

In my belief, the best or perhaps only way of testing this conjecture is by using this "right angle algorithm" on a triangle $P$ to see if such a triangle $P'$ can be found.
